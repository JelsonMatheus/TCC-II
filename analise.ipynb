```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import griddata
import numpy as np
```

```python
df = pd.read_csv('Resultados/otimização_5_min.csv')
correlation_matrix = df.corr()

print("Matriz de correlação:")
correlation_matrix
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['initial_stop_percent'], df['retorno'])
plt.title('initial_stop_percent vs Retorno')
plt.xlabel('initial_stop_percent')
plt.ylabel('Retorno')
plt.grid(True)
plt.show()

# Gráfico de dispersão para stop_increment vs Retorno
plt.figure(figsize=(10, 6))
plt.scatter(df['stop_increment'], df['retorno'])
plt.title('stop_increment vs Retorno')
plt.xlabel('stop_increment')
plt.ylabel('Retorno')
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv('Resultados/otimização_5_min.csv')

X, Y = np.meshgrid(df['initial_stop_percent'], df['stop_increment'])
points = np.array(df[['initial_stop_percent', 'stop_increment']])
values = df['retorno']

xi, yi = np.meshgrid(np.linspace(df['initial_stop_percent'].min(), df['initial_stop_percent'].max(), 361),
                     np.linspace(df['stop_increment'].min(), df['stop_increment'].max(), 361))
Z = griddata(points, values, (xi, yi), method='cubic')

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('Relação entre as variável initial_stop_percent, stop_increment e retorno')
ax.set_xlabel('initial_stop_percent')
ax.set_ylabel('stop_increment')
ax.set_zlabel('retorno')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
plt.show()
```
